```python
plt.loglog(em.e*1000,em.sigma_pho)
plt.loglog(em.e*1000,em.sigma_inc)
plt.loglog(em.e*1000,em.sigma_pp)
plt.loglog(em.e*1000,em.sigma_att)
plt.ylim([1e-3,1e5])
plt.xlim([1e-1,1e5])
plt.xlabel('$E_{\gamma}$ (keV) ')
plt.ylabel('$\mu$ (cm$^{-1}$)')
```

```python
Eg = 500
ct = np.arange(-1,1,0.00001)
KN = em.KleinNishina(Eg,ct)
KNfree = em.KleinNishina(Eg,ct,formfactor=False)
Eg_prime = Eg * em.P(Eg,ct)
plt.plot(ct,KN)
plt.plot(ct,KNfree)

plt.ylim([0,10])
plt.xlim([0.9,1.])
plt.show()
plt.plot(ct,Eg-Eg_prime)
```
